```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.__version__
```

```python
edad_x = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65]
salario_cd_y = [20000, 30000, 33000, 40000, 42000, 45000,47000, 60000, 70000, 80000]
salario_fed_y = [10000, 15000, 17000, 27000, 30000, 32000, 35000, 40000, 49000, 50000]
salario_bed_y = [15000, 20000, 25000, 27000, 35000, 40000, 45000, 50000, 55000, 60000]
```

```python
plt.plot(edad_x, salario_cd_y, label = 'Cientifico de datos', linewidth=3, marker='.', markersize=10, linestyle='--', color='orangered')
plt.plot(edad_x, salario_fed_y, label = 'Front end')
plt.plot(edad_x, salario_bed_y, label = 'Back end')

plt.title('Salario Medio de programadores')
plt.xlabel('Edad')
plt.ylabel('Salario Medio')

plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
plt.plot(edad_x, salario_cd_y, label = 'Cientifico de datos')
plt.plot(edad_x, salario_fed_y, label = 'Front end')
plt.plot(edad_x, salario_bed_y, label = 'Back end')

salario_medio = np.sum([salario_bed_y, salario_cd_y, salario_fed_y], axis=0)/3
plt.plot(edad_x, salario_medio, label = 'Salario medio', linewidth=3, marker='.', markersize=10, linestyle='--', color='black' )


plt.title('Salario Medio de programadores')
plt.xlabel('Edad')
plt.ylabel('Salario Medio')

plt.grid(True)
plt.legend()
# guardar la imagen
#plt.savefig('salario_medio.png')
plt.show()
```

```python

plt.plot(edad_x, salario_bed_y, label = 'Back end')

salario_medio = np.sum([salario_bed_y, salario_cd_y, salario_fed_y], axis=0)/3
plt.plot(edad_x, salario_medio, label = 'Salario medio', linewidth=3, marker='.', markersize=10, linestyle='--', color='black' )

plt.fill_between(edad_x, salario_bed_y, salario_medio,where=(salario_bed_y > salario_medio), interpolate= True, alpha=0.33, label= 'Subre la media')
plt.fill_between(edad_x, salario_bed_y, salario_medio,where=(salario_bed_y <= salario_medio),color='orangered', interpolate= True, alpha=0.33, label= 'bajo la media')


plt.title('Salario Medio de programadores')
plt.xlabel('Edad')
plt.ylabel('Salario Medio')

plt.grid(True)
plt.legend()
# guardar la imagen
#plt.savefig('salario_medio.png')
plt.show()
```

```python
width = 0.1

edad_np_x = np.arange(len(edad_x))

plt.bar(edad_np_x - 2*width, salario_cd_y, label = 'Cientifico de datos', width=width)
plt.bar(edad_np_x, salario_fed_y, label = 'Front end',width=width)
plt.bar(edad_np_x + 2*width, salario_bed_y, label = 'Back end',width=width)
# Para barras horizontales poner plt.barh()

salario_medio_y = np.sum([salario_bed_y, salario_cd_y, salario_fed_y], axis=0)/3
plt.bar(edad_np_x + 4*width, salario_medio_y, label = 'Salario medio', color='black',width=2*width )


plt.title('Salario Medio de programadores')
plt.xlabel('Edad')
plt.ylabel('Salario Medio')

plt.xticks(ticks=edad_np_x, labels=edad_x)
plt.grid(True)
plt.legend()
# guardar la imagen
#plt.savefig('salario_medio.png')
plt.show()
```

```python

# Para barras horizontales poner plt.barh()
plt.barh(edad_x, salario_cd_y, label = 'Salario ciencia de datos', color='black' )


plt.title('Salario Medio de programadores')
plt.xlabel('Salario Medio')
plt.ylabel('Edad')

plt.grid(True)
plt.legend()
# guardar la imagen
#plt.savefig('salario_medio.png')
plt.show()
```

```python
plt.pie(salario_cd_y, labels= edad_x, shadow= True, wedgeprops={'edgecolor':'black'}, autopct= '%1.2f%%')
plt.title('Edad y Salario de C. de datos')
plt.show()
```
